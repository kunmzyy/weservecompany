```python
import pandas as pd
```

```python
df_call_details = pd.read_csv('call details.csv')
```

```python
df_call_log=pd.read_csv('call log.csv')
```

```python
def remove_special_charater(calltype):
    result= calltype.replace('-','')
    return result
```

```python
df_call_details['callType']=df_call_details['callType'].apply(remove_special_charater)
```

```python
df_call_details
```

```python
df_call_log
```

```python
df_call_log.shape
```

```python
df_call_log.info()
```

```python
df_call_log['status']=df_call_log['status'].str.lower()
```

```python
df_call_log[df_call_log['resolutionDurationInHours'].isna()]
```

```python
df_call_log['resolutionDurationInHours'] = df_call_log['resolutionDurationInHours'].fillna(0).astype('int64')
```

```python
df_call_log[df_call_log['assignedTo'].isna()]
```

```python
df_call_log['assignedTo']=df_call_log['assignedTo'].fillna('0').astype('int64')
```

```python
df_call_log['resolutionDurationInHours']=df_call_log['resolutionDurationInHours'].fillna('not given')
```

```python
df_call_log
```

```python
df_call_details.to_csv('cleaned_call_details.csv', index=False)
```

```python
df_call_log.to_csv('cleaned_call_log.csv', index=False)
```
